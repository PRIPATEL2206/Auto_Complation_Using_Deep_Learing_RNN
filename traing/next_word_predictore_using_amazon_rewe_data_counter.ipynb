```python
import tensorflow as tf
```

```python
import pandas as pd
df= pd.read_csv("../data/amazon_reviews.csv")
df.head()
```

```python
df[:2000]
```

```python
text="/n".join(df[:2000]["content"].tolist())
# final_data=[]
# for i in text.split("/n"):
#     final_data.append(i.split())
# final_data
```

```python
tokenizer=tf.keras.preprocessing.text.Tokenizer()
```

```python
tokenizer.fit_on_texts([text])
tokenizer.word_index
```

```python
input_sequence=[]
step=0
for sen in text.split("/n"):
    token_sen=tokenizer.texts_to_sequences([sen])[0]
    for i in range(1,len(token_sen)):
        input_sequence.append(token_sen[:i+1])
    print(step)
    step+=1
len(input_sequence)
```

```python
input_sequence
```

```python
input_sequence[]
```

```python
max_sen_lenth=max(map(lambda x : len(x),input_sequence))
```

```python
max_sen_lenth
```

```python
padded_input_sequence=tf.keras.preprocessing.sequence.pad_sequences(
    input_sequence,
    maxlen=max_sen_lenth,
    padding="pre"
)
```

```python
len(padded_input_sequence)
```

```python
padded_input_sequence.shape
```

```python
padded_input_sequence
```

```python
X=padded_input_sequence[:,:-1]
X
```

```python
y=padded_input_sequence[:,-1]
y
```

```python
num_of_token=len(tokenizer.word_counts)
```

```python
num_of_token
```

```python
y=tf.keras.utils.to_categorical(y,num_classes=num_of_token+1)
```

```python
model=tf.keras.Sequential([
    tf.keras.layers.Embedding(num_of_token+1,200),
     tf.keras.layers.LSTM(150),
     tf.keras.layers.Dense(num_of_token+1,activation="softmax")
    ])
```

```python
model.compile(
    loss="categorical_crossentropy",
    optimizer="adam",
    metrics=["accuracy"]
)
```

```python
model.summary()
```

```python
model.fit(X,y,epochs=5)
```

```python
text_text='Too expensive I '
test_vec=tokenizer.texts_to_sequences([text_text])[0]
pad_test_vec = tf.keras.preprocessing.sequence.pad_sequences( [test_vec],
    maxlen=max_sen_lenth,
    padding="pre")
```

```python
test_vec
```

```python
import numpy as np

pred_index=np.argmax(model.predict([pad_test_vec])[0])
pred_index

tokenizer.index_word.get(pred_index)
```
